```python
import pandas as pd
import zipfile
import os
import requests
import io

# ------------------- CONFIGURATION -------------------
# Namespace and repo (update as needed)
namespace = "pytorch"  # Example: "OWASP", "tensorflow"
repo = "pytorch"  # Example: "Nettacker", "kubernetes"

# File format selection (choose one: "feather", "parquet", "zip")
file_format = "feather"

# Base GitHub URL (update if needed)
base_url = f"https://github.com/shunte88/mvdata/raw/main/data/{namespace}_{repo}"

# ------------------- DATA LOADING -------------------
df_chunks = []

for chk in range(1, 20):  # Adjust as needed
    file_url = f"{base_url}/{repo}_issues_{str(chk).zfill(4)}.{file_format}"

    try:
        if file_format == "feather":
            df_chunks.append(pd.read_feather(file_url))
        elif file_format == "parquet":
            df_chunks.append(pd.read_parquet(file_url, engine="pyarrow"))
        elif file_format == "zip":
            response = requests.get(file_url)
            if response.status_code == 200:
                with zipfile.ZipFile(io.BytesIO(response.content)) as z:
                    for file in z.namelist():
                        if file.endswith(".csv"):
                            df_chunks.append(pd.read_csv(z.open(file)))
            else:
                break
    except Exception as e:
        print(f"Skipping {file_url}: {e}")
        break

# Merge all chunks
if df_chunks:
    df = pd.concat(df_chunks, ignore_index=True)
    print(f"Loaded {len(df)} records from {namespace}_{repo}")
else:
    print("No data loaded.")
    exit()

# ------------------- DATA CLEANING -------------------
# Convert date columns
date_columns = ["created_at", "updated_at", "closed_at"]
for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# Fill missing values
df.fillna({"state_reason": "unknown", "body": "No description", "labels": "No labels"}, inplace=True)

# Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Drop duplicates
df.drop_duplicates(inplace=True)

print("\nâœ… Data Cleaning Completed!")
df.info()

# ------------------- EXPLORATORY ANALYSIS -------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Issue state distribution
plt.figure(figsize=(8, 5))
sns.countplot(y=df["state"], order=df["state"].value_counts().index)
plt.title("Issue State Distribution")
plt.xlabel("Count")
plt.ylabel("State")
plt.show()

# Issue creation over time
if "created_at" in df.columns:
    plt.figure(figsize=(12, 6))
    df["created_at"].dt.date.value_counts().sort_index().plot(kind="line")
    plt.title("Issues Created Over Time")
    plt.xlabel("Date")
    plt.ylabel("Number of Issues")
    plt.xticks(rotation=45)
    plt.show()

# Top 10 Labels
if "labels" in df.columns:
    label_counts = df["labels"].explode().value_counts().head(10)
    plt.figure(figsize=(12, 6))
    sns.barplot(y=label_counts.index, x=label_counts.values)
    plt.title("Top 10 Most Common Issue Labels")
    plt.xlabel("Count")
    plt.ylabel("Labels")
    plt.show()

print("\nâœ… Exploratory Data Analysis Completed!")

# ------------------- MVB OUTPUT GENERATION -------------------
# Sample summarization logic
def summarize_issue(row):
    return f"{row['title']} ({row['state']}) - Opened by {row['author']} on {row['created_at'].date()}"

df["mvb_summary"] = df.apply(summarize_issue, axis=1)

# Generate MVB Brief
top_issues = df.sort_values("created_at", ascending=False).head(5)
mvb_brief = "\n".join(top_issues["mvb_summary"].tolist())

print("\nðŸ“ MVB Summary Output:")
print(mvb_brief)

# Save the output as a CSV
df.to_csv(f"{repo}_issues_cleaned.csv", index=False)
print(f"\nâœ… Cleaned data saved as {repo}_issues_cleaned.csv")
```
